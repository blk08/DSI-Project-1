5. Return the player's power value.

````python
# players=players dictionary created in question 5 with pokedex details of player's pokemon
# k= player's id
def power(players,k):
    player_power=[]
    # j outputs players pokemon id
    for j in players[k]['player_pokemon'].keys():
        # extracting value for pokemon attack value
        x_attack = int(players[k]['player_pokemon'][j]['attack']) 
        # extracting value for pokemon special attack value
        x_sp_attack = int(players[k]['player_pokemon'][j]['special attack'])
        # extracting value for pokemon defence value
        x_defence = int(players[k]['player_pokemon'][j]['defence'])
        # extracting value for pokemon special defense value
        x_sp_defence = int(players[k]['player_pokemon'][j]['special defense'])
        # summing all the base stats (attack and defense) of player pokemon to find the player power
        x_player_power=x_attack+x_sp_attack+x_defence+x_sp_defence
        # storing player power for each player pokemon
        player_power.append(x_player_power)
        print('Name:',players[k]['player_pokemon'][j]['name'],'Attack:',players[k]['player_pokemon'][j]['attack'],'Special Attack:',players[k]['player_pokemon'][j]['special attack'],'Defence:',players[k]['player_pokemon'][j]['defence'],'Special Defence:',players[k]['player_pokemon'][j]['special defense'])
     


    print('Player Name:',players[k]['name'],'Player Power:',sum(player_power))
        
    
y=power(players,2)
          

    
````

- Rows are separated by newline characters: \n
- Columns are separated by commas: ,
- All cells in the csv are double quoted. Ex: "PokedexNumber" is the first cell of the first row.


Using for-loops, create a list of lists where each list within the overall list is a row of the csv/matrix, and each element in that list is a cell in that row. Additional criteria:

1. Quotes are removed from each cell item.
2. Numeric column values are converted to floats.
3. There are some cells that are empty and have no information. For these cells put a -1 value in place.

Your end result is effectively a matrix. Each list in the outer list is a row, and the *j*th elements of list together form the *j*th column, which represents a data attribute. The first three lists in your pokedex list should look like this:

    ['PokedexNumber', 'Name', 'Type', 'Total', 'HP', 'Attack', 'Defense', 'SpecialAttack', 'SpecialDefense', 'Speed']

````python
# Code to read in pokedex info
raw_pd = ''
pokedex_file = 'pokedex_basic.csv'
with open(pokedex_file, 'r') as f:
    raw_pd = f.read()
    
# the pokedex string is assigned to the raw_pd variable
````

````python
clean1 = raw_pd.splitlines()
````

````python
clean1
````

````python
# for looping to convert the data from string to list
data_unclean=[]
for row in clean1:
    data_unclean.append(row.split(','))
# for looping to remove double quotations    
data_clean=[]
for row in data_unclean:
    temp_data=[]
    for item in row:
        temp_data.append(item.replace('"',''))
    data_clean.append(temp_data)
    
# for looping to change "" to '-1'
data_clean1=[]
for row in data_clean:
    temp_data=[]
    for item in row:
        if item=='':
            temp_data.append('-1')
        else:
            temp_data.append(item)
    data_clean1.append(temp_data)
        
# for looping to change string value to integer
data_clean_2=[]
for row in data_clean1:
    temp_data=[]
    for item in row:
        try:
            temp_data.append(float(item))
        except ValueError:
            temp_data.append(item)
    data_clean_2.append(temp_data)
print(data_clean_2[:3])
````

````python
def pokefunc(pokedex,filter_options):

    # finding out the number of filter parameters 
    total_filters=len(list(filter_options.keys()))

    filtered_pokemon=[]
    # accessing each pokemon in the pokedex
    for pokemon in pokedex.values():
        # starting a count to ensure that every filter parameters has been checked
        count=0
        # finding the key and value in the filter_options dictionary
        for key,value in filter_options.items():
        #print(key,value)
        
            # if condition for the filter parameter for type
            if type(value) is str:
                # if condition to determine pokemon that matches with the filter parameter for type
                if value==pokemon[key]:
                    # add one to count if that pokemon is the type in the filter parameter
                    count+=1
            # if pokemon satisfies first if condition, it then proceeds to the next condition
            else:
                # pokemon is check for the other filter parameters condition
                if pokemon[key]>=value:
                    # add one to count if that pokemon satisfies the filter parameter condition
                    count+=1
        # once the pokemon has satisfies all the filter parameter condition check if the count matches with the no. of filter parameters specified
        if count==total_filters:
            # once all filtered condition is satisfied, add pokemon to filterd list
            filtered_pokemon.append(pokemon)

    print(filtered_pokemon)

filter_options = {
    'Attack':   25,
    'Defense':  30,
    'Type':     'Electric'}
filtered1=pokefunc(pokedex,filter_options)

            
        
````

### 9.1

What is the population mean and standard deviation of the "Total" attribute for all characters in the Pokedex?



````python
# for looping to add all Total of pokemon in pokedex
total_att=[]
for key in pokedex.keys():
    total_att.append(pokedex[key]['Total'])
# population mean 
mean_population=sum(total_att)/len(total_att)
print('population mean is',mean_population)

# for looping to find (x-x_mean)^2
total_att_sq=[]
for i in total_att:
    total_att_sq.append((i-mean_population)**2)
# standard deviation
std_population=(sum(total_att_sq)/len(total_att))**0.5
print('standard deviation of population is',std_population)
    
    
    
````
